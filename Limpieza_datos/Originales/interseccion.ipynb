```python
def inhibidor_riguroso(IC50):
    if IC50 < 1000:
        return 1
    else:
        return 0
```

```python
inhibidores_rigurosos = merge_nuevo[(merge_nuevo["Actividad_EGFR"] == 1) & (merge_nuevo["Actividad_MAPK1"] == 1) & (merge_nuevo["Actividad_SRC"] == 1)]
```

```python
inhibidores_rigurosos.to_csv("Inhibidores_rigurosos.csv")
```
